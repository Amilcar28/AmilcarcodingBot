Outras maneiras de criar arrays

```python
#imprimido com array funcao zero
dadoszero2=np.zeros((2,3))
dadoszero2
```

```python
# Criando um dados de sequencia array de 20 digitos separado de 2 em 2.
dseq=np.arange(0,20,2)
dseq
```

```python
#Transformar os dados array em float64
dseqf=dseq.astype(np.float64)
dseqf
```
