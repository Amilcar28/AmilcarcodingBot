```python
#2.Leia os dados, concatena e salve.

time = iris.Constraint(time=lambda cell: PartialDateTime(year=1964)
                            <= cell.point <= PartialDateTime(year=2022))

infile = os.path.join(DATADIR, 'ERA5_*.nc')
data = iris.load(infile, time)

print(data)

print('*' * 120)

unify_time_units(data)

equalise_attributes(data)
         
cube = data.concatenate_cube()

print('*' * 120)

outfile = os.path.join(DATADIR, 'ERA5_All.nc')
iris.save(cube, outfile)
print('Saved: {}'.format(outfile))
```

```python
def get_quarter(month):
    quarter_dictionary = {
        "Q1" : [1,2,3],
        "Q2" : [2,3,4],
        "Q3" : [3,4,5],
        "Q4" : [5,6,7],
        "Q5" : [6,7,8],
        "Q6" : [9,10,11],
        "Q7" : [10,11,12],
        "Q8" : [11,12,1],
        "Q9" : [12,2,1]
    }

    for key,values in quarter_dictionary.items():
        for value in values:
            if value == month:
                return key

print(get_quarter(3))
```
